```python
import matplotlib.pyplot as plt
```

```python
def logistic(r):
  n  = 0.1
  for i in range(1000):
    n = r * n * (1.0 - n)
    if i > 990:
      print(n)
```

```python
logistic(3.1)
```

```python
def logistic_plot(start, end, x, y):
  for i in range(1000):
    r = (end - start) * i / 1000 + start
    n = 0.1
    for j in range(1000):
      n = r * n * (1.0 - n)
      if j > 900:
        x.append(r)
        y.append(n)
```

```python
def plot(start, end):
  x, y = [], []
  logistic_plot(start, end, x, y)
  plt.scatter(x, y, s=0.1)
```

```python
plot(1.0, 3.0)
```

```python
plot(1.0, 4.0)
```

```python
plt.ylim([0.335,0.39])
plot(3.54, 3.58)
```
