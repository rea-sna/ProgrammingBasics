```python
from collections import defaultdict
import re
import io
import urllib.request
from zipfile import ZipFile
```

```python
def load_from_url(url):
  data = urllib.request.urlopen(url).read()
  zipData = ZipFile(io.BytesIO(data))
  fileName = zipData.namelist()[0]
  text = zipData.read(fileName).decode("shift-jis")
  text = re.sub(r'［.*?］', '', text)
  text = re.sub(r'《.*?》', '', text)
  return text
```

```python
URL = "https://www.aozora.gr.jp/cards/000119/files/624_ruby_5668.zip"
text = load_from_url(URL)
text.split()[0]
```

```python
def show_top10(text):
  m = MeCab.Tagger()
  node = m.parseToNode(text)
  dic = defaultdict(int)
  while node:
    a = node.feature.split(",")
    key = node.surface
    if a[0] == u"名詞" and a [1] == u"一般" and key != "":
      dic[key] += 1
      node = node.text
  for k, v in sorted(dic.items(), key=lambda x: -x[1])[0:10]:
    print(k + ":" + str(v))
```

```python
URL = "https://www.aozora.gr.jp/cards/000119/files/624_ruby_5668.zip"
text = load_from_url(URL)
show_top10(text)
```

```python
import IPython
from wordcloud import WordCloud
```

```python
def get_words(text):
  w = ""
  m = MeCab.Tagger()
  node = m.parseToNode(text)
  while node:
    a = node.feature.split(",")
    if a[0] == u"名詞" and a[1] == u"一般":
      w += node.surface + " "
    node  = node.next
  return w
```

```python
URL = "https://www.aozora.gr.jp/cards/000081/files/1935_ruby_19924.zip"
text = load_from_url(URL)
words = get_words(text)
fpath='/usr/share/fonts/opentype/ipafont-gothic/ipagp.ttf'
wc = WordCloud(background_color="white", width=1920, height=1080, font_path=fpath)
wc.generate(words)
wc.to_file("wc.png")
IPython.display.Image(filename="wc.png")
```
